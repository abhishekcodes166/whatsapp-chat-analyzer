```python
import re
import pandas as pd
```

```python
f = open("_chat.txt",'r', encoding='utf-8')
data = f.read()
```

```python
print(data)
```

```python
pattern = r"\[\d{1,2}/\d{1,2}/\d{2,4},\s*\d{1,2}:\d{2}:\d{2}\s*[AP]M\]"
```

```python
messages = re.split(pattern, data)[1:]  
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'date': dates, 'message': messages})
df["date"] = pd.to_datetime(df["date"], format="[%d/%m/%y, %I:%M:%S %p]")

df.head()
```

```python
df.shape
```

```python
 #seperate users and messages
users = []
messages_only = []
for message in df['message']:
    if ": " in message:
        user, msg = message.split(": ", 1)
        users.append(user)
        messages_only.append(msg)
    else:
        users.append("System")
        messages_only.append(message)
df['user'] = users
df['message_only'] = messages_only
df.head()
```

```python
#delete message column
df = df.drop(columns=['message'])
```

```python
df.head()
```

```python
df['year'] = df['date'].dt.year
```

```python
df.head()
```

```python
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df.head()
```
